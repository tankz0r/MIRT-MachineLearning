# Экспоненциальное распределение

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Генерируем выборку

```python
expon_rv= sts.expon(0)
sample = expon_rv.rvs(1000)
```

Строим гистограмму выборки и рисуем теоретическую плотность

```python
plt.hist(sample, normed=True, histtype='stepfilled', alpha=0.2)
plt.ylabel('number of samples')
plt.xlabel('$x$')
x = np.linspace(sts.expon.ppf(0.01),sts.expon.ppf(0.999), 1000)
pdf = expon_rv.pdf(x)
plt.plot(x, pdf,lw=5,label='theoretical pdf')
plt.legend(loc='best', frameon=False)
```

Генерируем 1000 выборок объемом 5

```python
a=np.zeros([1000,5])
for i in xrange(1000):
        a[i]=expon_rv.rvs(5)
e=np.zeros([1000,1])
for i in xrange(1000):
    for j in xrange(5):
        e[i]=e[i]+a[i,j]
    e[i]=e[i]/5
```

Вычисляем мат.ожидание и дисперсию

Строим гистограмму распределений выборочных средних и плотность нормального распределения

Посчитаем теоретически N(Ex,Dx/n)
expon.pdf(x) = exp(-x) с этого делаем вывод, что lambda=1
Для экспоненциального распределения E=1/lambda D=1/(lambda)^2, тогда E=1 D=1

```python
plt.hist(e, normed=True, histtype='stepfilled', alpha=0.2)
plt.ylabel('Empiric average')
plt.xlabel('$x$')
E=1
D=1
norm_rv = sts.norm(E,0.2)
x = np.linspace(-4,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
```
